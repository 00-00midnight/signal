## Гармонический сигнал

```python
import matplotlib.pyplot as mpl
import numpy

for i in range (0, 4):
    result = []

    for t in numpy.arange(0, 1.001, 0.001):
        result.append(numpy.sin(2*numpy.pi*2**i*t))
    
    
    f, (signal, spectrum) = mpl.subplots(2, 1)
    f.set_figheight(8)
    f.set_figwidth(20)
    signal.grid()
    spectrum.grid()
    spectrum.set_xticks(range(0, 51))
    spectrum.set_axisbelow(True)
    
    signal.plot((0, 1), (0, 0), c='g')
    signal.plot(numpy.arange(0, 1.001, 0.001), result, c='b', linewidth=3)
    signal.title.set_text(f"частота {2**i} герц")
    
    x = numpy.fft.rfftfreq(1001, 0.001)
    y = numpy.fft.rfft(result)
    
    spectrum.bar(x[0:51], abs(y[0:51])/500, width=0.5)
```

# Однополярный меандр


```python
import matplotlib.pyplot as mpl
import numpy

def module(x):
    if x < 0:
        return 0
    else:
        return x

for i in range (0, 4):
    result = []
    
    for t in numpy.arange(0, 1.001, 0.001):
        result.append(module(numpy.sign(numpy.sin(2*numpy.pi*2**i*t))))
    
    f, (signal, spectrum) = mpl.subplots(2, 1)
    f.set_figheight(8)
    f.set_figwidth(20)
    signal.grid()
    spectrum.grid()
    spectrum.set_xticks(range(0, 51))
    spectrum.set_axisbelow(True)
    
    signal.plot((0, 1), (0, 0), c='g')
    signal.plot(numpy.arange(0, 1.001, 0.001), result, c='b', linewidth=3)
    signal.title.set_text(f"частота {2**i} герц")
    
    x = numpy.fft.rfftfreq(1001, 0.001)
    y = numpy.fft.rfft(result)
    
    y[0] = 0
    spectrum.bar(x[0:51], abs(y[0:51])/500, width=0.5)
```
